```python
df_temp=pd.get_dummies(df,columns=['airline','source_city','departure_time','stops','arrival_time','destination_city','class'])
df_temp.drop(columns=['flight'],inplace=True)
correlations=df_temp.corr()
```

```python
top_5_corr = correlations['price'].drop('price').sort_values(ascending=False).head(5)
print("TOP 5 POSITIVELY CORRELATED ATRIBUTES TO PRICE\n",top_5_corr)

lowest_5_corr=correlations['price'].drop('price').sort_values(ascending=False).tail(5)
print("TOP 5 NEGATIVELY CORRELATED ATRIBUTES TO PRICE\n",lowest_5_corr)
```

```python
print("\n Most Expensive Flights ")
print(df.nlargest(5, 'price')[['airline', 'class', 'duration', 'price']])

print("\n Cheapest Flights ")
print(df.nsmallest(5, 'price')[['airline', 'class', 'duration', 'price']])
```

```python
df['route'] = df['source_city'] + " to " + df['destination_city']
top_routes = df['route'].value_counts().head(5)
plt.figure(figsize=(10, 6))
top_routes.plot(kind='barh')
plt.title('Top 5 Busiest Routes')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(24,18))
plt.subplot(3,3,1)
ax=sns.countplot(x=df['stops'],palette='pastel')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distribution Of The No. Of Stops Of  Flight')

plt.subplot(3,3,2)
ax=sns.countplot(x=df['departure_time'],palette='pastel')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distribution Of Departure Times for Flight')


plt.subplot(3,3,3)
ax=sns.countplot(x=df['class'],palette='pastel')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distribution Of Flight Classes ')


plt.subplot(3,3,4)
ax=sns.countplot(x=df['arrival_time'],palette='pastel')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distribution Of Flight Arrival Time')


plt.subplot(3,3,5)
ax=sns.countplot(x=df['source_city'],palette='pastel')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distribution Of Source Cities Of Flight')


plt.subplot(3,3,6)
ax=sns.countplot(x=df['destination_city'],palette='pastel')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distribution Of Destination Cities Of Flight')
plt.show()
```

```python
for col in ['airline', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'class']:
    sns.histplot(df[col], bins=20, kde= True, edgecolor='black', color='skyblue')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='airline', y='price', data=df)
plt.title('Price Distribution by Airline')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='stops', y='price', data=df)
plt.title('Price Distribution by Stop')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='class', y='price', data=df)
plt.title('Price Distribution by class')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='days_left', y='price', data=df)
plt.title('Price Trend vs Days Before Departure')
plt.gca().invert_xaxis()  
plt.show()
```
